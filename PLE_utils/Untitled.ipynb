```python
import numpy as np
from os.path import dirname, join as pjoin
import scipy.io as sio
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.signal import find_peaks
%matplotlib notebook
```

```python
PLEnum = 42553
filename = 'PLtlb1piezoitr_' + str(PLEnum)+'.mat'
data_dir = pjoin('/Users/zihuaiz/Desktop/zihuaiz/Python/Data Processing/PLE_utils', filename)
```

```python
mat_contents = sio.loadmat(data_dir)
```

```python
mat_contents['pl'].shape
```

```python
fig, ax = plt.subplots(2,1,figsize=(6, 6))
X = mat_contents['xvals'][0]
Y = mat_contents['yvals'][0]
Y = np.linspace(1,len(Y),len(Y))
Z = mat_contents['pl'].transpose()
ax[0].pcolormesh(X,Y,Z,shading = 'auto',cmap=cm.inferno)
ax[0].set_yticks(Y)
ax[1].pcolormesh(X,Y,mat_contents['ref'].transpose(),shading = 'auto',cmap=cm.inferno)
ax[1].set_yticks(Y)
for i in range(len(mat_contents['yvals'][0])):
    peakthreshold = 0.005
    xx = mat_contents['xvals'][0]
    ref = -mat_contents['ref'].transpose()[i,:]
    peaks, _ = find_peaks(ref,height = peakthreshold)
    loc = np.ones(len(peaks))*(i+1)
    ax[1].plot(xx[peaks],loc,'x')
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
peakthreshold = 0.005
xx = mat_contents['xvals'][0]
ref = -mat_contents['ref'].transpose()[0,:]
peaks, _ = find_peaks(ref,height = peakthreshold)
ax.plot(xx[peaks],ref[peaks],'x')
ax.plot(xx,ref)
```
